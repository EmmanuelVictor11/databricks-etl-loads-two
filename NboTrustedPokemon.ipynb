```
df = spark.sql("""
select * from dbraw.tbpokemon
""")
```

```
dfTrusted = spark.sql("""
    SELECT
    md5(concat(coalesce(Nome,'-'),
               coalesce(altura, '-'),
               coalesce(Peso, '-' ),
               coalesce(habilidades, '-' )))                                         as id_chave
        ,CAST(Nome                                                   as string)      as ch_Nome
        ,CAST(Altura                                                 as float)       as ch_Altura
        ,CAST(Peso                                                   as float)       as ch_Peso
        ,CAST(Habilidades                                            as string)      as ch_Habilidades
        ,CAST(Experiencia_Base                                       as float)       as ch_Experiencia_Base
        ,CAST(regexp_extract(Status, 'special-attack=([0-9]+)', 1)   as int)         as vl_status_special_attack
        ,CAST(regexp_extract(Status, 'defense=([0-9]+)', 1)          as int)         as vl_status_defense
        ,CAST(regexp_extract(Status, 'attack=([0-9]+)', 1)           as int)         as vl_status_attack
        ,CAST(regexp_extract(Status, 'hp=([0-9]+)', 1)               as int)         as vl_status_hp
        ,CAST(regexp_extract(Status, 'special-defense=([0-9]+)', 1)  as int)         as vl_status_special_defense
        ,CAST(regexp_extract(Status, 'speed=([0-9]+)', 1)            as int)         as vl_status_speed
        ,CAST(tipos                                                  as string)      as ch_tipos
        ,dt_referencia_criacao                                                       
        ,dt_data_atualizacao                                                         dt_referencia_atualizacao
    FROM dbraw.tbpokemon
""")
dfTrusted.createOrReplaceTempView('vw_pokemon')
```

```
(dfTrusted
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", True)
    .save("/dbfs/tmp/dbtrusted/pokemon")
)
     
```

```
spark.sql("""
    CREATE TABLE IF NOT EXISTS dbtrusted.tbpokemon USING DELTA LOCATION '/dbfs/tmp/dbtrusted/pokemon'
""")
```
